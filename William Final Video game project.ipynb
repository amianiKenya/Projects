```python
#ELEMENT 4
wil_line=np.polyfit(Wil_games.Year, Wil_games.JP_Sales,1)# defines our lint of best fit in the first order polynomial
wil_line_fit=wil_line[0]*Wil_games.Year+wil_line[1]#fits the line to the data that we have in the graph to be plotted.
```

```python
plt.plot(Wil_games.Year, Wil_games.JP_Sales, '*',color="brown")#plots the graph of global sales against year
plt.plot(Wil_games.Year, wil_line_fit)
print("The rate of change of sales per year in Japan is", wil_line[0], "which is a slow negative rate." )
plt.xlabel("Year") # labels y axis
plt.ylabel("sales in millions"); #labels x axis 
```

Japan had a wonderful market for video game  ever since but it has started declining due to some reasons.

```python
wil_line=np.polyfit(Wil_games.Year, Wil_games.EU_Sales,1)# defines our lint of best fit in the first order polynomial
wil_line_fit=wil_line[0]*Wil_games.Year+wil_line[1]#fits the line to the data that we have in the graph to be plotted.
plt.plot(Wil_games.Year, Wil_games.EU_Sales, '*')#plots the graph of EU sales against year
plt.plot(Wil_games.Year, wil_line_fit)#plots the line of best fit
print("The rate of change of sales per year in Europe is", wil_line[0],"and this a slow positive rate of change over the years.")
plt.xlabel("Year") # labels y axis
plt.ylabel("sales in millions"); #labels x axis 
```

```python
wil_line=np.polyfit(Wil_games.Year, Wil_games.Global_Sales,1) #define the fit line in the first order polynomial
wil_line_fit=wil_line[0]*Wil_games.Year+wil_line[1] #fits the line to the data that we have in the graph to be plotted.
plt.plot(Wil_games.Year, Wil_games.Global_Sales, '*')#plots the graph of global sales against year
plt.plot(Wil_games.Year, wil_line_fit)#plots the line of best fit
print("The rate of change of sales per year globally is", wil_line[0],"and this a slow negative rate of change over the years.")
#the above line is a print statement  
plt.xlabel("Year") # labels y axis
plt.ylabel("sales in millions"); #labels x axis 
```
